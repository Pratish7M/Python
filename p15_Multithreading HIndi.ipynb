```python
'''Write a program that spawns two threads. One thread should print even numbers between
 0 and 10,
 and the other thread should print odd numbers between 1 and 9.'''

import threading

def print_even_numbers():
  for i in range(0, 11, 2):
    print(i)

def print_odd_numbers():
  for i in range(1, 10, 2):
    print(i)

# Create threads
thread1 = threading.Thread(target=print_even_numbers)
thread2 = threading.Thread(target=print_odd_numbers)

# Start threads
thread1.start()
thread2.start()

# Wait for threads to finish (optional)
thread1.join()
thread2.join()

print("Finished printing numbers")
```

```python
''' Q63. Write a program that simulates a bank account transaction. The program should create
 two threads,
 one for a deposit and one for a withdrawal. The deposit thread should add
 100 to the account,and the withdrawal thread should withdraw'''
import threading

class BankAccount:
    def __init__(self):
        self.balance = 0
        self.lock = threading.Lock()  # Corrected: Lock object creation

    def deposit(self, amount):
        with self.lock:
            self.balance += amount  # Corrected: Use += for addition
            print(f"Deposited ${amount}. Current balance: ${self.balance}")

    def withdraw(self, amount):
        if amount <= self.balance:
            with self.lock:
                self.balance -= amount  # Corrected: Use -= for subtraction
                print(f"Successfully withdrew ${amount}. Current balance: ${self.balance}")
        else:
            print("Insufficient funds")

def deposit_thread(account, amount):
    account.deposit(amount)

def withdraw_thread(account, amount):
    account.withdraw(amount)

if __name__ == "__main__":
    account = BankAccount()
    deposit_amount = 500
    withdraw_amount = 100

    deposit_thread = threading.Thread(target=deposit_thread, args=(account, deposit_amount))
    withdraw_thread = threading.Thread(target=withdraw_thread, args=(account, withdraw_amount))

    deposit_thread.start()
    withdraw_thread.start()

    deposit_thread.join()
    withdraw_thread.join()

    print(f"Final balance: ${account.balance}")  # Corrected: Consistent formatting
```
